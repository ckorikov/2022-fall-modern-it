–†–∞—Å—à–∏—Ñ—Ä–æ–≤–∫–∞:
*   SUV - –∫—Ä–æ—Å—Å–æ–≤–µ—Ä -—Ç–∏–ø –∞–≤—Ç–æ–º–æ–±–∏–ª—è, –ø—Ä–æ–∏—Å—Ö–æ–¥—è—â–∏–π –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ –∏–∑ –∫–ª–∞—Å—Å–∞ –≤–Ω–µ–¥–æ—Ä–æ–∂–Ω–∏–∫–æ–≤, –Ω–æ –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã–π –≤ –ø–æ–¥–∞–≤–ª—è—é—â–µ–º –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–µ —Å–ª—É—á–∞–µ–≤ –≤ –≥–æ—Ä–æ–¥–∞—Ö –∏ –Ω–∞ –∞—Å—Ñ–∞–ª—å—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –¥–æ—Ä–æ–≥–µ;

*   Pickup - –∞–≤—Ç–æ–º–æ–±–∏–ª—å —Å –∑–∞–∫—Ä—ã—Ç–æ–π –∫–∞–±–∏–Ω–æ–π –∏ –æ—Ç–∫—Ä—ã—Ç–æ–π –≥—Ä—É–∑–æ–≤–æ–π –ø–ª–æ—â–∞–¥–∫–æ–π;
*   Coupe -  —Ç–∏–ø –∑–∞–∫—Ä—ã—Ç–æ–≥–æ –∫—É–∑–æ–≤–∞ –ª–µ–≥–∫–æ–≤–æ–≥–æ –∞–≤—Ç–æ–º–æ–±–∏–ª—è, –∫–æ—Ç–æ—Ä—ã–π –∏–º–µ–µ—Ç –¥–≤–µ –¥–≤–µ—Ä–∏, –æ–¥–∏–Ω –∏–ª–∏ –¥–≤–∞ —Ä—è–¥–∞ —Å–∏–¥–µ–Ω–∏–π;

*   Liftback - —Ä–∞–∑–Ω–æ–≤–∏–¥–Ω–æ—Å—Ç—å —Ö—ç—Ç—á–±–µ–∫–∞. –ì–ª–∞–≤–Ω–æ–µ –æ—Ç–ª–∏—á–∏–µ ¬´–ª–∏—Ñ—Ç–∞¬ª –æ—Ç ¬´—Ö—ç—Ç—á–∞¬ª –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ —Ç–æ–º, —á—Ç–æ —É –ø–µ—Ä–≤–æ–≥–æ –µ—Å—Ç—å –Ω–∞ –∑–∞–¥–Ω–µ–π –¥–≤–µ—Ä–∏ ¬´—Å—Ç—É–ø–µ–Ω—å–∫–∞¬ª, –∫–æ—Ç–æ—Ä–∞—è –≤ –ø—Ä–æ—Ñ–∏–ª—å –¥–µ–ª–∞–µ—Ç –∞–≤—Ç–æ–º–æ–±–∏–ª—å –ø–æ—Ö–æ–∂–∏–º –Ω–∞ —Å–µ–¥–∞–Ω;
*   Hatchback - –∫—É–∑–æ–≤ –ª–µ–≥–∫–æ–≤–æ–≥–æ –∞–≤—Ç–æ–º–æ–±–∏–ª—è —Å –æ–¥–Ω–∏–º –∏–ª–∏ –¥–≤—É–º—è —Ä—è–¥–∞–º–∏ —Å–∏–¥–µ–Ω–∏–π, –¥–≤–µ—Ä—å—é –≤ –∑–∞–¥–Ω–µ–π —Å—Ç–µ–Ω–∫–µ –∏ —É–∫–æ—Ä–æ—á–µ–Ω–Ω—ã–º –∑–∞–¥–Ω–∏–º —Å–≤–µ—Å–æ–º;
*   Roadster - –¥–≤—É—Ö–º–µ—Å—Ç–Ω—ã–π –∞–≤—Ç–æ–º–æ–±–∏–ª—å —Å –º—è–≥–∫–æ–π –∏–ª–∏ –∂—ë—Å—Ç–∫–æ–π —Å—ä—ë–º–Ω–æ–π –∫—Ä—ã—à–µ–π
*   Compact MPV - –∫–æ–º–ø–∞–∫—Ç–Ω—ã–π –º–∏–Ω–∏–≤—ç–Ω;
*   Hardtop Coupe - –∑–∞–∫—Ä—ã—Ç—ã–π –∫—É–∑–æ–≤ –ª–µ–≥–∫–æ–≤–æ–≥–æ –∞–≤—Ç–æ–º–æ–±–∏–ª—è, –≤ –∫–æ—Ç–æ—Ä–æ–º –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω—ã–µ —Å—Ç–æ–π–∫–∏ –∫—Ä—ã—à–∏;
*   Van - —Ç–∏–ø –∑–∞–∫—Ä—ã—Ç–æ–≥–æ –∞–≤—Ç–æ–º–æ–±–∏–ª—å–Ω–æ–≥–æ –∫—É–∑–æ–≤–∞, –ø—Ä–µ–¥–Ω–∞–∑–Ω–∞—á–µ–Ω–Ω—ã–π –¥–ª—è –ø–µ—Ä–µ–≤–æ–∑–∫–∏ –≥—Ä—É–∑–æ–≤, –∏–Ω–æ–≥–¥–∞ —Å –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å—é –ø–µ—Ä–µ–≤–æ–∑–∫–∏ –ø–∞—Å—Å–∞–∂–∏—Ä–æ–≤;
*   Limousine - –∑–∞–∫—Ä—ã—Ç—ã–π –∫—É–∑–æ–≤ –ª–µ–≥–∫–æ–≤–æ–≥–æ –∞–≤—Ç–æ–º–æ–±–∏–ª—è —Å –∂—ë—Å—Ç–∫–æ–π –∫—Ä—ã—à–µ–π, –æ—Å–Ω–∞—â—ë–Ω–Ω–æ–π –ø–æ–¥—ä—ë–º–Ω—ã–º —Å—Ç–µ–∫–ª–æ–º, –æ–±—ã—á–Ω–æ c –ø–µ—Ä–µ–≥–æ—Ä–æ–¥–∫–æ–π –º–µ–∂–¥—É –æ—Ç–¥–µ–ª–µ–Ω–∏–µ–º –≤–æ–¥–∏—Ç–µ–ª—è –∏ –æ—Å—Ç–∞–ª—å–Ω—ã–º —Å–∞–ª–æ–Ω–æ–º;
*   Convertible - r–∞–±—Ä–∏–æ–ª–µ—Ç - —Ç–∏–ø –∫—É–∑–æ–≤–∞ –ª–µ–≥–∫–æ–≤–æ–≥–æ –∞–≤—Ç–æ–º–æ–±–∏–ª—è —Å –º—è–≥–∫–∏–º –æ—Ç–∫–∏–¥–Ω—ã–º –∏–ª–∏ –∂–µ—Å—Ç–∫–∏–º —Å—ä–µ–º–Ω—ã–º –≤–µ—Ä—Ö–æ–º; 
*   Microvan - –º–∏–∫—Ä–æ–≤—ç–Ω (–≤ –æ—Å–Ω–æ–≤–Ω–æ–º —è–ø–æ–Ω—Å–∫–∏–µ) - –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –º–∏–∫—Ä–æ–≤—ç–Ω–æ–≤ –∏–º–µ—é—Ç –¥–≤–µ –ø–µ—Ä–µ–¥–Ω–∏–µ –¥–≤–µ—Ä–∏, –¥–≤–µ —Å–∫–æ–ª—å–∑—è—â–∏–µ –¥–≤–µ—Ä–∏ –∏ –±–æ–ª—å—à—É—é –∑–∞–¥–Ω—é—é;
*   Fastback - —Å–æ–±–∏—Ä–∞—Ç–µ–ª—å–Ω–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ç–∏–ø–æ–≤ –∞–≤—Ç–æ–º–æ–±–∏–ª—å–Ω—ã—Ö –∫—É–∑–æ–≤–æ–≤, –∏–º–µ—é—â–∏—Ö –ø–æ–∫–∞—Ç—É—é —Ñ–æ—Ä–º—É –∫—Ä—ã—à–∏, –ø–ª–∞–≤–Ω–æ, –±–µ–∑ —Å—Ç—É–ø–µ–Ω—å–∫–∏, –ø–µ—Ä–µ—Ö–æ–¥—è—â–µ–π –≤ –∫—Ä—ã—à–∫—É –±–∞–≥–∞–∂–Ω–∏–∫–∞;

```python
def get_cols(df) -> list:
    '''
    function return list of name numbers and categorials columns
    '''
    categorical_feature_mask = df.dtypes == object
    number_feature_mask = df.dtypes != object
    numbers_cols = df.columns[number_feature_mask].tolist()
    categorical_cols = df.columns[categorical_feature_mask].tolist()
    return [numbers_cols, categorical_cols]

numbers_cols, categorical_cols = get_cols(df)
for column in categorical_cols:
  columne_most_freq = df[column].value_counts()[:10].sort_values(ascending=False).index
  plt.title(f'{column} vs Log Price')
  plt.figure(figsize=(10,10))
  ax = sns.boxplot(
            x=df[df[column].isin(columne_most_freq)][column], 
            y=np.log10(df[df[column].isin(columne_most_freq)].Price), 
            order=columne_most_freq
            )
  plt.xticks(rotation=45)
```
