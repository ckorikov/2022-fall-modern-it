# –õ–∏–Ω–µ–π–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫, –ø–æ–∫–∞–∑—ã–≤–∞—é—â–∏–π –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–∏–¥–µ–æ–∏–≥—Ä, –≤—ã–ø—É—â–µ–Ω–Ω—ã—Ö –ø–æ –∂–∞–Ω—Ä–∞–º —Å —Ç–µ—á–µ–Ω–∏–µ–º –≤—Ä–µ–º–µ–Ω–∏

2) –ò–∑–º–µ–Ω–µ–Ω–∏—è –≤ —Ç–µ—Ö–Ω–æ–ª–æ–≥–∏—è—Ö: 

–í–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ –∏–≥—Ä–æ–≤—ã—Ö —Ç–µ—Ö–Ω–æ–ª–æ–≥–∏–π —Ç–∞–∫–∂–µ –º–æ–≥—É—Ç –≤–ª–∏—è—Ç—å –Ω–∞ –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∂–∞–Ω—Ä–æ–≤. –ù–∞–ø—Ä–∏–º–µ—Ä, —Ä–∞—Å—Ç—É—â–∏–π —Ä–µ–∞–ª–∏–∑–º –∏ –ø–æ–≥—Ä—É–∂–µ–Ω–∏–µ –∫–æ–Ω—Å–æ–ª–µ–π –º–æ–≥—É—Ç —Å–¥–µ–ª–∞—Ç—å –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–µ –∂–∞–Ω—Ä—ã, —Ç–∞–∫–∏–µ –∫–∞–∫ –ø—Ä–∏–∫–ª—é—á–µ–Ω—á–µ—Å–∫–∏–µ –∏–≥—Ä—ã –∏–ª–∏ —à—É—Ç–µ—Ä—ã, –±–æ–ª–µ–µ –ø—Ä–∏–≤–ª–µ–∫–∞—Ç–µ–ª—å–Ω—ã–º–∏ –¥–ª—è –∏–≥—Ä–æ–∫–æ–≤.

# –õ–∏–Ω–µ–π–Ω—ã–µ –∏ —Å—Ç–æ–ª–±—á–∞—Ç—ã–µ –¥–∏–∞–≥—Ä–∞–º–º—ã, –ø–æ–∫–∞–∑—ã–≤–∞—é—â–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ä—ã–Ω–∫–∞ –≤–∏–¥–µ–æ–∏–≥—Ä –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º –≤–æ –≤—Ä–µ–º–µ–Ω–∏

```python
global_df = df.groupby(['Year'])["Global"].sum().reset_index(name='sum')
global_df["Region"] = "Global"
japan_df = df.groupby(['Year'])["Japan"].sum().reset_index(name='sum')
japan_df["Region"] = "Japan"
na_df = df.groupby(['Year'])["North America"].sum().reset_index(name='sum')
na_df["Region"] = "North America"
europe_df = df.groupby(['Year'])["Europe"].sum().reset_index(name='sum')
europe_df["Region"] = "Europe"
row_df = df.groupby(['Year'])["Rest of World"].sum().reset_index(name='sum')
row_df["Region"] = "Rest of World"

line_chart_df = global_df.append(japan_df)
line_chart_df = line_chart_df.append(na_df)
line_chart_df = line_chart_df.append(europe_df)
line_chart_df = line_chart_df.append(row_df).reset_index()

fig = px.line(line_chart_df, x="Year", y="sum", color='Region', markers=True)
fig.show()
```

```python
bar_chart_df = line_chart_df[line_chart_df["Region"] != "Global"]
fig = px.bar(bar_chart_df, x="Year", y="sum", color='Region', text_auto=True)
fig.show()
```
