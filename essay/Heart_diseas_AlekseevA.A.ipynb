Атрибут принимает 2 значения: 0 = менее  120 мг/дл ; 1 = более  120 мг/дл

*   ca(number of major vessels (0-3) colored by flourosopy) - количество крупных сосудов (0-3), окрашенных при флюороскопии
*   thal(thal: 0 = normal; 1 = fixed defect; 2 = reversable defect) - Результат нагрузочного теста талия (0 = NA; 1 = Фиксированный дефект; 2 = Нормальный; 3 = Обратимый дефект)
*   target - этот атрибут показывает наличие болезни сердца у пациента (целочисленное значение 0 = отсутствие болезни и 1 = болезнь)




# Медицинские термины

*Стенокардия* — клинический синдром, характеризуемый ощущением дискомфорта за грудиной. Появляется внезапно при физической нагрузке или эмоциональном стрессе.

*   *Бессимптомная стенокардия* - ишемия миокарда, как правило, документально подтвержденная в стресс-тесте, не сопровождается болью в груди или эквивалентами стенокардии. (Значение атрибута cp = 0)

*   *Типичная стенокардия* (Значение атрибута cp = 3)

1.   Наличие типичной боли или дискомфорта в загрудинной области, продолжающейся от 3 до 5 минут
2.   Провоцируется физическим или эмоциональным стресом
3.   Купируется в покое или нитроглицирином в течение нескольких минут


*   *Атипичная стенокардия* определяется наличием двух из приведённых выше критериев (Значение атрибута cp = 1)
*   *Неангинальная боль* определяется наличием лишь одного из трёх критериев либо боль не отвечает ни одному из критеривев, указанных для типичной стенокардии (Значение атрибута cp = 2)

*Кровяное давление* - давление, которое кровь оказывает на стенки кровеносных сосудов, иначе говоря, превышение давления жидкости в кровеносной системе над атмосферным. 
Нормальное давление:

*   129/80 – для людей, младше 65 лет
*   139/80 – для пациентов старшего возраста

*Холестерин* – это вещество, из которого в организме человека формируются атеросклеротические бляшки. Именно они являются причиной проявления атеросклероза, являющегося очень опасной болезнью.

*   Менее 200 мг/дл — оптимальный уровень холестерина
*   От 200 мг/для до 239 мг/дл — максимально допустимый уровень
*   От 240 мг/дл и выше — критически высокий уровень холестерина

*Сегмент ST* – это отрезок кривой ЭКГ, анализируемый в целях диагностики заболеваний

*ЭКГ* в состоянии покоя
*   0 = гипертрофия левого желудочка (явное увеличение массы миокарда левого желудочка)
*   1 = нормальная
*   2 = аномалия волн ST-T
