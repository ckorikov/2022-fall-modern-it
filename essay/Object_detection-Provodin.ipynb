Определяем свойства изображения, которое мы открыли в программе, с помощью функций cv2. Затем строим подзаголовок изображения, используя функцию subplot() plt и задавая в ней параметры. Функции imshow() и show() модуля plt, чтобы показать изображение на выходе.

На выходе изображение отображается в результате работы программы


```python
# Import OpenCV module 
import cv2 
import matplotlib
# Import pyplot from matplotlib as plt 
from matplotlib import pyplot as pltd 
# Opening the image from files 
face_cascade=cv2.CascadeClassifier(cv2.data.haarcascades + "XML-data.xml") #Note the change

imaging = cv2.imread("stop.png") 
# Altering properties of image with cv2 
imaging_gray = cv2.cvtColor(imaging, cv2.COLOR_BGR2GRAY) 
imaging_rgb = cv2.cvtColor(imaging, cv2.COLOR_BGR2RGB) 

# Importing Haar cascade classifier xml data 

xml_data = cv2.CascadeClassifier('XML-data.xml') 
# Detecting object in the image with Haar cascade classifier  
detecting = xml_data.detectMultiScale(imaging_gray,  
                                   minSize =(80, 80)) 
# Amount of object detected 
amountDetecting = len(detecting) 
# Using if condition to highlight the object detected 
if amountDetecting != 0: 
    for(a, b, width, height) in detecting: 
        cv2.rectangle(imaging_rgb,(a, b), # Highlighting detected object with rectangle 
                     (a + height, b + width),  
                     (0, 275, 0), 9) 
# Plotting image with subplot() from plt 
pltd.subplot(1, 1, 1) 
# Displaying image in the output 
pltd.imshow(imaging_rgb) 
pltd.show() 
```

Используем функцию detectMultiScale() в программе для обнаружения объекта.

Импорт в программу XML-файл каскадного классификатора. Затем используем функцию detectMultiScale() с импортированным каскадным файлом, чтобы определить, присутствует ли объект на изображении или нет.

В условии if в программе проверяем обнаружен ли объект или нет, и если объект обнаружен, мы выделили часть обнаруженного объекта, используя цикл for с функциями cv2. Выделив на изображении часть обнаруженного объекта, мы отобразили обработанное изображение с помощью функций plt show() и imshow().
